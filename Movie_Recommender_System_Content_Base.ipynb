Now Convert string to list

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
import ast

def convert(text):
  l = []
  for i in ast.literal_eval(text):
    l.append(i['name'])
  return l
```

```python
# movies['genres'].apply(convert)
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(2)
```

```python
movies.iloc[0]['keywords']
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
import ast

def convert_cast(text):
  l = []
  counter = 0
  for i in ast.literal_eval(text):
    if counter < 3:
      l.append(i['name'])
    counter += 1
  return l
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies.head(2)
```

```python
movies.iloc[0]['crew']
```

```python
import ast

def fetch_director(text):
  l = []

  for i in ast.literal_eval(text):
    if i['job'] == 'Director':
      l.append(i['name'])
      break
  return l
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head(2)
```

```python
movies.iloc[0]['overview']
```

```python
type(movies.iloc[0]['overview'])
```

Concatinating
eg : Sam Worthington to SamWorthington

```python
def remove_space(word):
  l = []
  for i in word:
    l.append(i.replace(" ",""))
  return l
```

```python
movies['cast'] = movies['cast'].apply(remove_space)
movies['crew'] = movies['crew'].apply(remove_space)
movies['genres'] = movies['genres'].apply(remove_space)
movies['keywords'] = movies['keywords'].apply(remove_space)
```

converting all tags to lower case lettrs

**Stem Operation**
eg: 'love', 'loved', 'loving' are same words. need to converts to one word 'love'

```python
import nltk
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stems(text):
  l = []
  for i in text.split():
    l.append(ps.stem(i))

  return " ".join(l)
```

```python
new_df['tags'] = new_df['tags'].apply(stems)
```

```python
new_df.iloc[0]['tags']
```

Now text is ready to use. Next need to perform counter vectorizer
and remove stop words

When you use this cv object to fit and transform text data, it will:

* Split the text into tokens (words or phrases).
* Count the occurrences of each token in each document.
* Filter out stop words.
* Keep only the top 5000 most frequent tokens.

# Perform CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
vector
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
similarity.shape
```

```python
new_df[new_df['title']=='Spider-Man'].index[0]
```

```python
def recommend(movie):
  index = new_df[new_df['title']== movie].index[0]
  distances = sorted(list(enumerate(similarity[index])), reverse = True, key = lambda x: x[1])
  for i in distances[1:6]:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('Spider-Man')
```

```python
recommend('Spectre')
```
